```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
with open('features.txt') as f:
    input_features = f.readlines()

input_features = [feature.strip() for feature in input_features]
input_features
```

```python
df = pd.read_csv("data_processed.csv")
df.head()
```

```python
import functools
def merge_review(series):
    return functools.reduce(lambda x, y: (str(x)+" "+str(y)), series)
```

```python
aggregate_data = df.groupby(["product_name"])["user_rating","product_review"].agg({"user_rating":"mean","product_review":merge_review})
aggregate_data.sort_values(by = "user_rating", ascending= False, inplace= True)
aggregate_data
```

```python
for feature in input_features:
    aggregate_data[feature] = aggregate_data["product_review"].apply(lambda x: x.count(feature))
aggregate_data
```

```python
aggregate_data["sum"] = aggregate_data[input_features].apply(lambda x: x.sum(), axis=1)
aggregate_data
```

## Approach - 1

```python
tmp = aggregate_data.sort_values(by = "sum", ascending= False)#[:10].sort_values(by="user_rating", ascending= False)[:3]
recommendation = tmp[(tmp['aroma'] >= tmp['aroma'].quantile(q=0.8)) & \
(tmp['taste'] >= tmp['taste'].quantile(q=0.8)) & \
(tmp['flavor'] >= tmp['flavor'].quantile(q=0.8))].sort_values(by="user_rating", ascending= False)
recommendation
```

```python
print("Recommended Beers are:")
for index, beer in recommendation[:3].iterrows():
    print("\t",index,"; Score:", round(beer["user_rating"],2))
```

## Approach - 2

```python
from numpy.linalg import norm

def projection(vec):
    vec = np.array(vec)
    base = np.array([[1],[1],[1]])/np.sqrt(3)
    return round(np.dot(vec,base)[0],5)

def get_deviation(vec):
    cosine = projection(vec)/norm(vec)
    return round(np.arccos(np.clip(cosine, -1, 1)),5)

recommendation = aggregate_data.copy()
recommendation["similarity"] = recommendation[input_features].apply(lambda x: projection(np.array(x)), axis = 1)
recommendation["proximity"] = recommendation[input_features].apply(lambda x: get_deviation(np.array(x)), axis = 1)
recommendation = recommendation.sort_values(by="similarity", ascending = False)[:5].sort_values("user_rating", ascending= False)
# tmp = tmp[:10]
recommendation
```

```python
print("Recommended Beers are:")
for index, beer in recommendation[:3].iterrows():
    print("\t",index,"; Score:", round(beer["user_rating"],2))
```
